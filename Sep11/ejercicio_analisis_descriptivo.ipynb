# Respondiendo a las preguntas respecto al censo de matrimonio y divorcios

Se obtuvieron los datos de https://github.com/resbaz/data/tree/master/marriage

```python
import os
import pandas as pd
import requests
```

```python
file = 'men.csv'

if not os.path.isfile(file):
    url = 'https://github.com/resbaz/data/blob/master/marriage/men.csv'
    html = requests.get(url).content
    men_list = pd.read_html(html)
    men = men_list[0]
    men.to_csv('men.csv')
else:
    men = pd.read_csv(file)

men
```

```python
file = 'women.csv'

if not os.path.isfile(file):
    url = 'https://github.com/resbaz/data/blob/master/marriage/women.csv'
    html = requests.get(url).content
    women_list = pd.read_html(html)
    women = women_list[0]
    women.to_csv('women.csv')
else:
    women = pd.read_csv(file)

women
```

```python
file = 'both_sex.csv'

if not os.path.isfile(file):
    url = 'https://github.com/resbaz/data/blob/master/marriage/both_sexes.csv'
    html = requests.get(url).content
    both_sex_list = pd.read_html(html)
    both_sex = both_sex_list[0]
    both_sex.to_csv('both_sex.csv')
else:
    both_sex = pd.read_csv(file)

both_sex
```

```python
file = 'divorce.csv'

if not os.path.isfile(file):
    url = 'https://github.com/resbaz/data/blob/master/marriage/divorce.csv'
    html = requests.get(url).content
    divorce_list = pd.read_html(html)
    divorce = divorce_list[0]
    divorce.to_csv('divorce.csv')
else:
    divorce = pd.read_csv(file)

divorce
```

Una vez obtenidas todas las tablas con datos del censo de matrimonios y divorcios, podemos comenzar a responder a las preguntas formuladas en clase.

1.- ¿Cómo es el porcentaje de matrimonios contra el procentaje de divorcios a lo largo de los años? ¿Hay algún año en el que los divorcios sean más que los matrimonios?

```python
#Tomamos los datos necesarios de las tablas de matrimonios (para ambos sexos) y divorcios

all_marriage = both_sex[['year', 'all_2534']]
all_marriage
```

```python
all_divorces = divorce[['year', 'all_3544']]
all_divorces
```

```python
marriage_vs_divorce = pd.merge(all_marriage, all_divorces)
marriage_vs_divorce
```

```python
marriage_vs_divorce_new = marriage_vs_divorce.rename(columns = {'year': 'Year', 'all_2534': 'Marriages', 'all_3544': 'Divorces'})
marriage_vs_divorce_new
```

Una inspección rápida parece índicar que el porcentaje de gente en matrimonios es siempre más alto que el de gente divorciada. Usando una gráfica

```python
marriage_vs_divorce_new.plot(x = 'Year', y = ['Marriages', 'Divorces'], kind = "bar")
```

3.- De acuerdo a los datos, ¿existe un grupo de personas que se case más acorde a su escolaridad? ¿Qué sucede con los divorcios?

```python
school = both_sex[['year', 'HS_2534', 'SC_2534', 'BAp_2534', 'BAo_2534', 'GD_2534']]
school
```

Separe la info en dos grupos dada la cantidad de diferentes opciones. Grafico por separado
