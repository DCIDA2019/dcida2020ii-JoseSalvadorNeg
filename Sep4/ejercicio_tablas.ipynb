```python
file = 'pobreza.csv'

if not os.path.isfile(file):
    url = 'https://www.indexmundi.com/map/?t=0&v=69&r=af&l=en'
    html = requests.get(url).content
    pobreza_lista = pd.read_html(html)
    pobreza = pobreza_lista[1]
    pobreza.to_csv('pobreza.csv')
else:
    pobreza = pd.read_csv(file)

pobreza
```

```python
file = 'desempleo_joven.csv'

if not os.path.isfile(file):
    url = 'https://www.indexmundi.com/map/?t=0&v=2229&r=af&l=en'
    html = requests.get(url).content
    desempleo_joven_lista = pd.read_html(html)
    desempleo_joven = desempleo_joven_lista[1]
    desempleo_joven.to_csv('desempleo_joven.csv')
else:
    desempleo_joven = pd.read_csv(file)

desempleo_joven
```

```python
file = 'desempleo.csv'

if not os.path.isfile(file):
    url = 'https://www.indexmundi.com/map/?t=0&v=74&r=af&l=en'
    html = requests.get(url).content
    desempleo_lista = pd.read_html(html)
    desempleo = desempleo_lista[1]
    desempleo.to_csv('desempleo.csv')
else:
    desempleo = pd.read_csv(file)

desempleo
```

```python
file = 'fuerza.csv'

if not os.path.isfile(file):
    url = 'https://www.indexmundi.com/map/?t=0&v=72&r=af&l=en'
    html = requests.get(url).content
    fuerza_lista = pd.read_html(html)
    fuerza = fuerza_lista[1]
    fuerza.to_csv('fuerza.csv')
else:
    fuerza = pd.read_csv(file)

fuerza
```

```python
#De forma alternativa con loc

pobreza.loc[:'1', :'Population below poverty line (%)']
```

```python
#Supongamos que solo nos interesan aquellos países con porcentaje de pobreza superior al 50 porciento, usamos

pobreza.loc[pobreza['Population below poverty line (%)'] > 50, ['Country', 'Population below poverty line (%)']]
```

```python
#En el caso de poseer un dato equivocado, es posible modificarlo. Digamos que el porcentaje de pobreza en
#Zimbabwe es del 80 porciento en vez del 72, usamos

pobreza.iloc[0, 1] = 80

pobreza.iloc[:1, :2]
```

```python
#Para obtener una tabla que contenga países con el mismo porcentaje de pobreza

pobreza.loc[pobreza['Population below poverty line (%)'] == 70, ['Country', 'Population below poverty line (%)']]
```

```python
#Hasta ahora hemos trabajado solo con la tabla de pobreza, pero podemos hacer uso del resto de tablas
#uniéndolas mediante el uso de merge

crisis = pd.merge(fuerza, desempleo)
crisis

#donde se ha fusionado los datos de la tabla de desempleo por la parte derecha
```

```python
#Podemos ignorar el año mediante

crisis_sin = pd.merge(fuerza, desempleo).drop('Year', axis = 1)
crisis_sin

#dando como resultado una tabla sin datos innecesarios
```
