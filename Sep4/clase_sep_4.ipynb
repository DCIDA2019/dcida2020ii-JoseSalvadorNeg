```python
file = 'pobreza.csv'

if not os.path.isfile(file):
    url = 'https://www.indexmundi.com/map/?t=0&v=69&r=af&l=en'
    html = requests.get(url).content
    pobreza_lista = pd.read_html(html)
    pobreza = pobreza_lista[1]
    pobreza.to_csv('pobreza.csv')
else:
    pobreza = pd.read_csv(file)

pobreza
```

```python
file = 'desempleo_joven.csv'

if not os.path.isfile(file):
    url = 'https://www.indexmundi.com/map/?t=0&v=2229&r=af&l=en'
    html = requests.get(url).content
    desempleo_joven_lista = pd.read_html(html)
    desempleo_joven = desempleo_joven_lista[1]
    desempleo_joven.to_csv('desempleo_joven.csv')
else:
    desempleo_joven = pd.read_csv(file)

desempleo_joven
```
