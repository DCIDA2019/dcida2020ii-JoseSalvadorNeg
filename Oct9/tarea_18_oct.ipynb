## Poisson

CDF: $\frac{\Gamma([K + 1], \lambda)}{[k]!}$ con $\Gamma$ la upper incomplete gamma function y $[k]!$ es la floor function

Mean: $\lambda$

Sigma: $\sqrt{\lambda}$

## Binomial

PDF: $\frac{n!}{k!(n - k)!} p^{k} q^{n - k}$ con $p$ probabilidad de xito, $q$ probabilidad de fracaso, $n$ nmero de intentos y $k$ nmero de xitos

CDF: $\sum_{i = 0}^{[K]} \frac{n!}{i!(n - i)!} p^{i} (1 - p)^{n - i}$

Mean: $np$

Sigma: $\sqrt{npq}$

```python
mean = 5
sigma = np.sqrt(2.5)

#n = 10, p = 0.5 y q = 0.5

datos_k = []
datos_bino = []
k = 0
while k <= 10:
    pdf = ((np.math.factorial(10))/(np.math.factorial(k)*np.math.factorial(10 - k)))*(0.5**k)*(0.5**(10 - k))
    k += 1
    datos_k.append(k)
    datos_bino.append(pdf)

plt.plot(datos_k, datos_bino, '-.', color = 'red')
plt.figure(figsize=(50,30))
plt.show()
```

4.- show explicitly that the linear least square method is derived from the minimization of the chi-square when the model is a straight line

Para una linea recta, el mtodo de chi-square es $\chi^{2} = \sum (y_{i} - a x_{i} - b)^{2}/ \sigma_{y_{i}}^{2}$. Debe minimizarse este factor a travs de sus derivadas respecto a los parmetros $a$ y $b$. Esto es $\frac{\partial \chi^{2}}{\partial a} = 0$ y $\frac{\partial \chi^{2}}{\partial b} = 0$. Las derivadas completas son

$\frac{\partial \chi^{2}}{\partial a} = \sum 2 (y_{i} - a x_{i} - b) (- x_{i}) / \sigma_{y_{i}}^{2} = 0$ $\hspace{10pt}$ y $\hspace{10pt}$
$\frac{\partial \chi^{2}}{\partial b} = \sum 2 (y_{i} - a x_{i} - b) (-1) / \sigma_{y_{i}}^{2} = 0$

O de otra forma

$\sum (y_{i} - a x_{i} - b) x_{i} = 0$ $\hspace{10pt}$ y $\hspace{10pt}$ $\sum (y_{i} - a x_{i} - b) = 0$

Depejamos $b$ de la segunda ecuacin como $nb = \sum (y_{i} - a x_{i})$ y sustituimos en la primer ecuacin

$\sum (y_{i} x_{i} - a x_{i}^{2} - [\frac{1}{n} \sum (y_{i} - a x_{i})] x_{i}) = 0$

$n \sum y_{i} x_{i} - n a \sum x_{i}^{2} - \sum y_{i} \sum x_{i} + a (\sum x_{i})^{2} = 0$

Despejar en $a$ nos lleva a

$a = \frac{n \sum x_{i} y_{i} - \sum y_{i} \sum x_{i}}{n \sum x_{i}^{2} - (\sum x_{i})^{2}}$

Sustituir esto en la expresin de $b$ nos lleva a

$b = \frac{\sum y_{i} \sum x_{i}^{2} - \sum x_{i} \sum x_{i} y_{i}}{n \sum x_{i}^{2} - (\sum x_{i})^{2}}$

Estos son los coeficientes correspondientes al least square linear method
